```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
academi_sub = pd.read_csv("sample_submission.csv")
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv")
```

```python
#missing valuesはないとサイトに記載されているが念のため確認
train.isna().sum() #train.isna()
```

```python
#Labelエンコーディング
from sklearn.preprocessing import LabelEncoder

train = pd.DataFrame(train)

# LabelEncoder のインスタンスを作成してターゲット変数を数値にエンコードする
encoder = LabelEncoder()
train['target_encoded'] = encoder.fit_transform(train['Target'].values)

print(train["target_encoded"])

#graduate:2, dropout:0, enroll:1

```

```python
#不均衡なデータかもしれない。ラベルの少ないデータの学習が不十分になる可能性
value_counts = train['target_encoded'].value_counts()

print(value_counts)
```

学習

```python
from sklearn.model_selection import train_test_split
#特徴量とラベルに分割
X=train.drop(["target_encoded","Target"],axis=1 )

Y=train["target_encoded"]

#データの分割
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2)


```

```python
X.head(5)
```

```python

Y_train=np.array(Y_train)
X_train=np.array(X_train)
```

```python
#OneVsRestClassifier

from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# モデル作成
clf = OneVsRestClassifier(LogisticRegression(max_iter=1000,random_state=0))
clf.fit(X_train, Y_train)

# モデル推論

# テストデータで予測を行う
Y_pred = clf.predict(X_test)

# 精度を計算する
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)

```

イテレーションを増やしたが、収束していない。データのスケーリングが必要？

```python
#ランダムフォレスト
from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(n_estimators = 500, random_state = 0) #500の決定木
clf.fit(X_train, Y_train)
print(clf.score(X_test, Y_test))
```

```python
# 交差検証を実行
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X, Y, cv=5)  # cvは分割数を指定

# 各分割のスコアを出力
print("Cross-validation scores:", scores)

# 平均スコアを出力
print("Average cross-validation score:", np.mean(scores))
```

```python
#ランダムフォレスト
from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(random_state = 42) #デフォルトの決定木の数
clf.fit(X_train, Y_train)
print(clf.score(X_test, Y_test))
```
