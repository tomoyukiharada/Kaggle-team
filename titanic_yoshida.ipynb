##モデル作成

```python
X = train.drop(columns = ["PassengerId", "Survived", "Name", "Cabin", "Ticket"])
Y = test.drop(columns = ["PassengerId",  "Name", "Cabin", "Ticket"])
```

```python
X = pd.get_dummies(X, columns=["Sex", "Embarked"], drop_first=True)
Y = pd.get_dummies(Y, columns=["Sex", "Embarked"], drop_first=True)
```

```python
X = X.drop(columns={"Sex_female"})
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, train["Survived"], test_size=0.2)
```

```python
X_train.shape[0] == Y_train.shape[0]
X_test.shape[0] == Y_test.shape[0]
```

```python
params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'binary_logloss',
    'num_leaves': 64,
    'min_data_in_leaf': 20,
    'max_depth': 7,
    'verbose': 0,
}
```

```python
import lightgbm as lgb
```

```python
gbm = lgb.LGBMClassifier(objective='binary')
```

```python
gbm.fit(X_train, Y_train)
```

```python
pre = gbm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pre, Y_test)
```
