##モデル作成

```python
X_ = train.drop(columns = ["PassengerId", "Survived", "Name", "Cabin", "Ticket"])
Y_ = test.drop(columns = ["PassengerId",  "Name", "Cabin", "Ticket"])
```

```python
X_ = pd.get_dummies(X_, columns=["Sex", "Embarked"], drop_first=True)
Y_ = pd.get_dummies(Y_, columns=["Sex", "Embarked"], drop_first=True)
```

```python
import lightgbm as lgb
```

```python
gbm = lgb.LGBMClassifier(objective='binary')
```

```python
from sklearn.model_selection import train_test_split
X, X_test, Y, Y_test = train_test_split(X_, train["Survived"], test_size=0.2)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

model_accuracy = []
models = []
preds = []

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
for fold, (train_indices, valid_indices) in enumerate(kf.split(X, Y)):
  x_train, x_valid = X.iloc[train_indices], X.iloc[valid_indices]
  y_train, y_valid = Y.iloc[train_indices], Y.iloc[valid_indices]


  model= gbm.fit(x_train, y_train)
  pre = gbm.predict(x_valid)
  print(accuracy_score(pre, y_valid))

  models.append(model)

for model in models:
  pred = gbm.predict_proba(X_test)
  preds.append(pred)

y_pred = np.mean(preds, axis=0)
print(f"score: {accuracy_score(y_pred.argmax(axis=1), Y_test)}")







```
